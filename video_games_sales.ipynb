```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
video_games_sales=pd.read_csv('vgsales.csv',)
video_games_sales
```

```python
#check for null values
video_games_sales.isna().any()
#check for any duplicates
video_games_sales.duplicated().any()
```

```python
#Columns with NULL values
video_games_sales[['Year','Publisher']]
```

```python
#Remove null values from csv
df_cleaned=video_games_sales.dropna()
#double check for any null values
df_cleaned.isna().any()
```

```python
#dropping unnecessary column 'Other_Sales' from dataframe
df_new_cleaned=df_cleaned.drop('Other_Sales',axis=1)
#print new data frame
df_new_cleaned
```

```python
#Grouping by Platform and the NA_Sales relating to the platform
platform_sales=df_new_cleaned.groupby('Platform')['NA_Sales'].sum().reset_index()
```

```python
#Displying 'Platform' and their 'NA_Sales'
platform_sales
```

```python
# Group less frequent platforms into 'Other'
top_N=10  # Number of top platforms to keep
top_selling_platforms=platform_sales.nlargest(top_N,'NA_Sales')
other_platforms = platform_sales[~platform_sales['Platform'].isin(top_selling_platforms['Platform'])]
other_sales = other_platforms['NA_Sales'].sum()
other_platforms = pd.DataFrame({'Platform': ['Other'], 'NA_Sales': [other_sales]})
# Concatenate top platforms and 'Other'
combined_platforms = pd.concat([top_selling_platforms, other_platforms])
# Create the bar plot
sns.barplot(data=combined_platforms, x='Platform', y='NA_Sales')
```

```python
genre_sales=df_new_cleaned.groupby('Genre')['NA_Sales'].sum().reset_index()
genre_sales
```

```python
# Group less frequent genres into 'Other'
top_ten=10 # Number of top platforms to keep
top_selling_genre=genre_sales.nlargest(top_ten,'NA_Sales')
other_genre=genre_sales[~genre_sales['Genre'].isin(top_selling_genre['Genre'])]
other_genre_sales=other_genre['NA_Sales'].sum()
# Concatenate top platforms and 'Other'
other_genre=pd.DataFrame({'Genre':['Other'], 'NA_Sales':[other_genre_sales]})
combined_genres = pd.concat([top_selling_genre, other_genre])
```

```python
# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(combined_genres['NA_Sales'], labels=combined_genres['Genre'], autopct='%1.1f%%', startangle=90)
plt.title('Top Selling Genres in NA')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()





```
